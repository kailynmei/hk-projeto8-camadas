### Parte 1 — Lado Emissor

#### 1.1 — Carregamento e Execução dos Áudios Originais

<div style="background:#fff0f7; color:black; border-left: 5px solid #e91e8c; padding: 14px 18px; border-radius: 6px; box-sizing: border-box; width: 100%;">

Carregue os 3 arquivos de áudio. Verifique a taxa de amostragem e execute cada áudio. Observe a forma de onda e a **transformada de Fourier** de cada sinal original, identificando a banda de frequências que cada um ocupa.

> **Atenção:** os sinais devem ser normalizados e, se necessário, convertidos para mono antes da modulação.

</div>

```python
fs = 44100  # Taxa de amostragem

# frequências das notas musicais 
notas = {
    'C4': 261.63, 'D4': 293.66, 'E4': 329.63, 'F4': 349.23,
    'G4': 392.00, 'A4': 440.00, 'B4': 493.88, 'C5': 523.25,
    'D5': 587.33, 'E5': 659.25, 'G5': 783.99, 'A5': 880.00,
    'REST': 0.0
}

def gerar_nota(freq, duracao, fs=44100, amplitude=0.5):
    """Gera uma nota com envelope ADSR suave para evitar cliques."""
    t = np.linspace(0, duracao, int(fs * duracao), endpoint=False)
    if freq == 0:
        return np.zeros_like(t)
    sinal = amplitude * np.sin(2 * np.pi * freq * t)
    # Envelope: fade-in e fade-out de 10ms
    fade = int(0.01 * fs)
    sinal[:fade] *= np.linspace(0, 1, fade)
    sinal[-fade:] *= np.linspace(1, 0, fade)
    return sinal

def gerar_melodia(sequencia, fs=44100):
    """Recebe lista de (nota, duracao_em_segundos) e retorna o sinal completo."""
    partes = [gerar_nota(notas[n], d, fs) for n, d in sequencia]
    return np.concatenate(partes)

# Melodia 1: Ode à Alegria 
seq1 = [
    ('E4',0.3),('E4',0.3),('F4',0.3),('G4',0.3),
    ('G4',0.3),('F4',0.3),('E4',0.3),('D4',0.3),
    ('C4',0.3),('C4',0.3),('D4',0.3),('E4',0.3),
    ('E4',0.45),('D4',0.15),('D4',0.6),
]

# Melodia 2: Parabéns pra Você
seq2 = [
    ('C4',0.3),('C4',0.15),('D4',0.45),('C4',0.45),('F4',0.45),
    ('E4',0.9),('C4',0.3),('C4',0.15),('D4',0.45),
    ('C4',0.45),('G4',0.45),('F4',0.9),
]

# Melodia 3: Twinkle Twinkle
seq3 = [
    ('C4',0.3),('C4',0.3),('G4',0.3),('G4',0.3),
    ('A4',0.3),('A4',0.3),('G4',0.6),
    ('F4',0.3),('F4',0.3),('E4',0.3),('E4',0.3),
    ('D4',0.3),('D4',0.3),('C4',0.6),
]

audio1 = gerar_melodia(seq1).astype(np.float32)
audio2 = gerar_melodia(seq2).astype(np.float32)
audio3 = gerar_melodia(seq3).astype(np.float32)

wavfile.write('audio1.wav', fs, audio1)
wavfile.write('audio2.wav', fs, audio2)
wavfile.write('audio3.wav', fs, audio3)

print(f"Áudio 1: {len(audio1)/fs:.2f}s  |  Áudio 2: {len(audio2)/fs:.2f}s  |  Áudio 3: {len(audio3)/fs:.2f}s")
```

```python
print("▶ Estação 1 — Ode à Alegria")
display(Audio(audio1, rate=fs))

print("▶ Estação 2 — Parabéns pra Você")
display(Audio(audio2, rate=fs))

print("▶ Estação 3 — Twinkle Twinkle")
display(Audio(audio3, rate=fs))
```

### • FFT dos áudios originais


```python
# FFT dos 3 áudios originais
fig, axs = plt.subplots(3, 1, figsize=(12, 8))

audios = [audio1, audio2, audio3]
titulos = ['Estação 1 — Ode à Alegria', 'Estação 2 — Parabéns pra Você', 'Estação 3 — Twinkle Twinkle']
cores = ['#e91e8c', '#ff6eb4', '#ffaad4']

for i, (audio, titulo, cor) in enumerate(zip(audios, titulos, cores)):
    N = len(audio)
    freqs = np.fft.rfftfreq(N, d=1/fs)
    magnitude = np.abs(np.fft.rfft(audio)) / N

    axs[i].plot(freqs, magnitude, color=cor, linewidth=0.8)
    axs[i].set_title(titulo, fontsize=11, fontweight='bold')
    axs[i].set_xlabel('Frequência (Hz)')
    axs[i].set_ylabel('Magnitude')
    axs[i].set_xlim(0, 1000)   
    axs[i].grid(True, alpha=0.3)

plt.suptitle('FFT dos Áudios Originais', fontsize=13, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```
